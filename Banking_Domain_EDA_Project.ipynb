```python
# Lets analyze the data in the Estimated Income column and lets group them according to the income
# Grp 1 = 0 to 20000
# Grp 1 = 20000 to 50000

# In this way we convert numerical data into categorical data

# Define income band boundaries
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Mid', 'High']

# Create the 'Income Band' column using pd.cut
df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, include_lowest=True)
```

Numerical Analysis

```python
# Numerical analysis and exploration
numerical_cols = ['Fee Structure','Age', 'Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

# Univariate analysis and visualization
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 4, i + 1)
    sns.histplot(df[col], kde=True)   # kde=True - Adds a smooth curved line over your histogram
    plt.title(col)
plt.tight_layout()
plt.show()
```

HeatMaps to check relation analysis

```python
# Select numerical columns for correlation analysis
numerical_cols = ['Age', 'Estimated Income', 'Superannuation Savings', 'Credit Card Balance',
                  'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts',
                  'Foreign Currency Account', 'Business Lending', 'Properties Owned']

# Calculate the correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

Pairing the plots to check relation between each column for detail analysis of the dataset.

```python
pairs_to_plot = [
    ('Bank Deposits', 'Saving Accounts'),
    ('Checking Accounts', 'Saving Accounts'),
    ('Checking Accounts', 'Foreign Currency Account'),
    ('Age', 'Superannuation Savings'),
    ('Estimated Income', 'Checking Accounts'),
    ('Bank Loans', 'Credit Card Balance'),
    ('Business Lending', 'Bank Loans'),
]


for x_col, y_col in pairs_to_plot:
    plt.figure(figsize=(8, 6))
    sns.regplot(
        data=df,
        x=x_col,
        y=y_col,
        scatter_kws={'alpha': 0.4},     # semi-transparent points
        line_kws={'color': 'red'}       # best-fit line color
    )
    plt.title(f'Relationship between {x_col} and {y_col}', fontsize=14)
    plt.xlabel(x_col, fontsize=12)
    plt.ylabel(y_col, fontsize=12)
    plt.tight_layout()
    plt.show()
```

Insights from the above analysis

Deposits and Savings Behavior
There is a strong link between Bank Deposits and Savings Accounts, which could mean they reflect similar financial habits—like how much money a person keeps in the bank. It also shows that people who deposit money regularly are likely to maintain or grow their savings.

Income, Age, and Accumulation
Age and Estimated Income show a moderate connection with balances in accounts like Superannuation, Savings, and Checking. This fits a common financial pattern—older people and those who earn more tend to save more, build retirement funds, and may also have higher credit card balances or loans.

Low Correlation with Properties Owned
Owning property seems less connected to these banking variables. That’s probably because property ownership depends on other things like location, the housing market, or inheritance—which aren’t reflected in this data.

Business vs. Personal Banking
There’s a moderate connection between Business Lending and Bank Loans, which suggests some people have both personal and business debts. But overall, business lending doesn’t relate much to deposits or property data. This shows it likely applies to a separate group of customers or needs.
